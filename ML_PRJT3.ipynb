# ***Printing the car won't start table ***

```python
var = ["bf", "no", "ng", "flb", "cs", "sb"]
val = ["y", "n"]
c1 = []
for bf in val:
  for no in val:
    for ng in val:
      for flb in val:
        for cs in val:
          for sb in val:
            c1.append((bf, no, ng, flb, cs, sb))

cws_table = {}
for c in c1:
  bf, no, ng, flb, sb, cs = c
  if "y" in [bf, no, ng, flb, sb]:
    cws_table[c] = 0.9 if cs == "n" else 0.1
  elif bf == "n" and no == "n" and ng == "n" and flb == "n" and sb == "n" and cs == "n":
    cws_table[c] = 0.1
  elif ('n', 'n', 'n', 'n', 'n', 'n') in cws_table:
    t = 1 - cws_table[(bf, no, ng, flb, sb, cs)]
    cws_table[c] = t
  else:
    t = 0
for c, p in cws_table.items():
  print(f"{c}: {p}")
```

# **R2: Computing the probability of P (-cs, +ab, +fb)**

```python
count = 0
total = 0
for c, p in cws_table.items(): #iterate
    bf, no, ng, flb, sb, cs = c
    if cs == "n" and bf == "y" and flb == "y":#satisfying the condition
        count += 1
        total += p
if count > 0: # probability
    prob = total / count
    print(f"P(-cs, +ab, +fb) = {prob}")
else:
    print("No such combination exists in the table.")
```
